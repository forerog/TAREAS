```python
nEdades=np.sort(np.unique(df_excel["edad"]))
iE=nEdades[0]
fE=nEdades[0]
n=len(nEdades)
print("n= ",nEdades[n-1])
for i in nEdades:
    fE+=4
    if(fE<nEdades[n-1]):
        print("Entre ",iE," y ",fE)
        print(df_excel.loc[(df_excel["edad"]>=iE) & (df_excel["edad"]<=fE),:]["precio"].sum())
    iE=fE
```

```python
#PUNTO 6
vusuarios=np.unique(df_excel["usuario"])
j=0
tTCompras=0
while(j < len(vusuarios)):
    mCompras=0
    tCompras=0
    con=0
    for i in range(len(df_excel["edad"])):
        if(df_excel["usuario"][i]==vusuarios[j]):
            tCompras+=df_excel["precio"][i]
            con+=1
    tTCompras+=tCompras
    mCompras=tCompras/con
    print(vusuarios[j])
    print("Recuento de compras ",con)
    print("Precio medio de compra ",round(mCompras,3))
    j+=1
print("Valor total de las compras ",round(tTCompras,3))

```

```python
#Punto 7
hugo=np.unique(df_excel["nombrearticulo"])
TotalCompras=[]
ConteoCompras=[]
PrecioArticulo=[]
IDarticulo=[]
for j in range(len(hugo)):
    mCompras=0
    tCompras=0
    con=0
    for i in range(len(df_excel["nombrearticulo"])):
      if(df_excel["nombrearticulo"][i]==hugo[j]):
        tCompras+=df_excel["precio"][i]
        con+=1
        if(con==1):
          PrecioArticulo.append(df_excel["precio"][i])
          IDart=df_excel["ID"][i]
    TotalCompras.append(tCompras)
    ConteoCompras.append(con)
    IDarticulo.append(IDart)
for i in range(len(hugo)):
  for j in range(len(hugo)):
    if(ConteoCompras[i] > ConteoCompras[j]):
      aux=TotalCompras[i] 
      TotalCompras[i]=TotalCompras[j]
      TotalCompras[j]=aux
      aux1=IDarticulo[i] 
      IDarticulo[i]=IDarticulo[j]
      IDarticulo[j]=aux1
      aux2=PrecioArticulo[i] 
      PrecioArticulo[i]=PrecioArticulo[j]
      PrecioArticulo[j]=aux2
      aux3=ConteoCompras[i] 
      ConteoCompras[i]=ConteoCompras[j]
      ConteoCompras[j]=aux3
      aux4=hugo[i] 
      hugo[i]=hugo[j]
      hugo[j]=aux4
k=0            
while(k<5):
    print("------------------------------------------------")
    print("ARTICULO N° ",k+1)
    print("------------------------------------------------")
    print("ID del artıculo ============> ",IDarticulo[k])
    print("Nombre del artıculo ========> ",hugo[k])
    print("Numero de compras ==========> ",ConteoCompras[k])
    print("Precio del artıculo ========> ",PrecioArticulo[k])
    print("Valor total de las compras => ",round(TotalCompras[k],2))
    print("------------------------------------------------")
    k+=1
```
